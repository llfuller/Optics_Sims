```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.ticker import FuncFormatter
import meep as mp
```

## Define Materials:
Define the optical properties of titanium dioxide (TiO2) and glass. MEEP uses relative permittivity (epsilon) to define material properties:

```python
TiO2 = mp.Medium(index=2.7)  # Example index, adjust based on actual data
glass = mp.Medium(index=1.5)
```

## Geometry Setup:
Define the geometry of your unit cell. Assume dimensions as needed or based on the project details:

```python
#  Choose dimensions that are multiple times larger than the wavelength to simulate periodicity correctly. For instance, 
# if simulating a single unit cell, you might choose sizes around a few micrometers, say 2μm.
x_size = 2.0 # micrometers
y_size = 2.0 # micrometers
#  This should be enough to include the material thickness plus 
# additional space for boundary layers or source and monitor placement. Assume a total of 2μm.
z_size = 2.0 # micrometers

# Material thicknesses
tio2_ridge_thickness = 0.1  # micrometers (100 nm)
tio2_base_thickness = 0.1  # micrometers (100 nm)
glass_base_thickness = 0.5  # micrometers (500 nm)

tio2_ridge_y_width = 0.1  # micrometers (100 nm)


cell_size = mp.Vector3(x_size, y_size, z_size)  # Define the cell size
geometry = [mp.Block(material=TiO2,
                     center=mp.Vector3(0, 0, glass_base_thickness + tio2_base_thickness + tio2_ridge_thickness/2),
                     size=mp.Vector3(-tio2_ridge_y_width/2, tio2_ridge_y_width/2, tio2_ridge_thickness)),
            mp.Block(material=TiO2,
                     center=mp.Vector3(0, 0, glass_base_thickness + tio2_base_thickness/2),
                     size=mp.Vector3(mp.inf, mp.inf, tio2_base_thickness)),
            mp.Block(material=glass,
                     center=mp.Vector3(0, 0, glass_base_thickness/2),
                     size=mp.Vector3(mp.inf, mp.inf, glass_base_thickness))]
```

```python
# # Plot dimensions
# plot_x = np.linspace(-x_size/2, x_size/2, 100)
# plot_y = np.linspace(-y_size/2, y_size/2, 100)
# plot_z = np.linspace(0, z_size, 100)
# Define the formatter function
def format_func(value, tick_number):
    labels = {0: 'Vacuum', 1: 'Glass', 2: 'TiO2'}
    return labels.get(value, '')

def material_map(geometry, y_range, z_range, resolution):
    # Initialize the material map with 0 (vacuum)
    y_points = np.linspace(y_range[0], y_range[1], resolution)
    z_points = np.linspace(z_range[0], z_range[1], resolution)
    map_matrix = np.zeros((len(z_points), len(y_points)))
    
    # Iterate over each point in the y-z plane
    for i, z in enumerate(z_points):
        for j, y in enumerate(y_points):
            # Check each block in the geometry
            for block in geometry:
                # Get the material type
                if block.material == TiO2:
                    material_type = 2
                elif block.material == glass:
                    material_type = 1
                else:
                    continue
                
                # Determine the bounds of the block in the y and z dimensions
                y_min = block.center.y - block.size.y/2
                y_max = block.center.y + block.size.y/2
                z_min = block.center.z - block.size.z/2
                z_max = block.center.z + block.size.z/2
                
                # Check if the current point (y, z) is within this block
                if y_min <= y <= y_max and z_min <= z <= z_max:
                    map_matrix[i, j] = material_type
                    break  # Stop checking other blocks if a match is found
    
    return map_matrix

# Define the range for y and z to be visualized and the resolution of the grid
y_range = (-1.0, 1.0)  # Adjust as needed
z_range = (-1.0, 2.0)  # Adjust as needed
resolution = 100

# Generate the material map
map_matrix = material_map(geometry, y_range, z_range, resolution)

# Optionally, you can visualize this map using matplotlib
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
im = plt.imshow(map_matrix, extent=[y_range[0], y_range[1], z_range[0], z_range[1]], origin='lower', aspect='auto')
cbar = plt.colorbar(im, ticks=[0, 1, 2])
cbar.ax.yaxis.set_major_formatter(FuncFormatter(format_func))
cbar.set_label('Material')
plt.xlabel('Y dimension')
plt.ylabel('Z dimension')
plt.title('Unit Cell Material Map in Y-Z Plane (Unbounded material in X and Y)')
plt.show()
```

## Source Definition:
Define an appropriate source, such as a plane wave

```python
freq = 5.45*10**14 #Hz (if light lambda= 550 nm, range 380-700nm)
pml_thickness = 0.5 # thickness of perfectly matched layer to absorb outgoing waves
sources = [mp.Source(mp.ContinuousSource(frequency=freq),
                     component=mp.Ez,
                     center=mp.Vector3(0,0,z_size-pml_thickness),
                     size=mp.Vector3(x_size, y_size, 0))]
```

## Boundary Conditions:
Typically, you'll use periodic boundary conditions for a unit cell:

```python
pml_layers = [mp.PML(pml_thickness)]
# boundary_layers = [mp.PeriodicBoundary(mp.Vector3(1,0,0), phase_shift=0),
#                    mp.PeriodicBoundary(mp.Vector3(0,1,0), phase_shift=0)]
```

## Simulation Domain:
Set up the simulation domain:

```python
sim = mp.Simulation(cell_size=cell_size,
                    geometry=geometry,
                    resolution = 10,
                    sources=sources)
```

## Run the Simulation:
Execute the simulation and monitor fields:


```python
sim.run(until=200)  # Adjust time as needed
```

## Visualize

```python
# import numpy as np
# import matplotlib.pyplot as plt
# from matplotlib.animation import FuncAnimation

def animate_field_2D(sim, duration, fps=10):
    # Prepare the figure and axis
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_title('Ez field over time')

    # Initialize a plot for Ez
    eps_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Dielectric)
    field_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Ez)
    z_index = field_data.shape[2] // 2  # Middle index along the z-axis
    img = ax.imshow(field_data[:, :, z_index], animated=True, extent=(-1,1,-1,1))

    # Set up colorbar once
    plt.colorbar(img, ax=ax)

    # Animation update function
    def update(frame):
        # Step the simulation
        sim.run(until=1)  # Update this as necessary for your timestep
        field_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Ez)
        img.set_data(field_data[:, :, z_index])
        img.set_clim(vmin=np.min(field_data), vmax=np.max(field_data))
        return (img,)

    # Create animation
    ani = FuncAnimation(fig, update, frames=int(duration*fps), interval=1000/fps, blit=True)
    plt.show()
    return ani

# Usage:
animate_field_2D(sim, duration=10)  # Duration is in seconds
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def animate_field_2D(sim, duration, fps=10):
    # Prepare the figure and axis
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_title('Ez field over time')

    # Initialize a plot for Ez
    eps_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Dielectric)
    field_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Ez)
    z_index = field_data.shape[2] // 2  # Middle index along the z-axis
    img = ax.imshow(field_data[:, :, z_index], animated=True, extent=(-1,1,-1,1))

    # Set up colorbar once
    plt.colorbar(img, ax=ax)

    # Animation update function
    def update(frame):
        # Step the simulation
        sim.run(until=1)  # Update this as necessary for your timestep
        field_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Ez)
        img.set_data(field_data[:, :, z_index])
        img.set_clim(vmin=np.min(field_data), vmax=np.max(field_data))
        return (img,)

    # Create animation
    ani = FuncAnimation(fig, update, frames=int(duration*fps), interval=1000/fps, blit=True)
    plt.show()
    return ani

# Usage:
final_animation = animate_field_2D(sim, duration=10)  # Duration is in seconds

# Saving the animation
final_animation.save('output.mp4', writer='ffmpeg', fps=10)
final_animation.save('output.gif', writer='imagemagick', fps=10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def animate_precomputed_field_2D(field_data_list, duration, fps=10):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_title('Ez field over time')

    # Assume first frame data to set up the plot
    img = ax.imshow(field_data_list[0], animated=True, extent=(-1, 1, -1, 1))
    plt.colorbar(img, ax=ax)

    # Update function for the animation
    def update(frame):
        img.set_data(field_data_list[frame])
        return (img,)

    # Total number of frames (make sure you don't exceed the list length)
    total_frames = int(duration * fps)
    if total_frames > len(field_data_list):
        total_frames = len(field_data_list)

    # Create the animation
    ani = FuncAnimation(fig, update, frames=total_frames, interval=1000/fps, blit=True)
    plt.show()
    return ani

# Usage
# Assume field_data_list is a list of numpy arrays, each representing the Ez field at a timestep
field_data_list = [sim.get_array(center=mp.Vector3(), size=mp.Vector3(2,2,2), component=mp.Ez) for timestep in range(200)]  # Adjust as needed
final_animation = animate_precomputed_field_2D(field_data_list, duration=10)  # Duration in seconds

# Save the animation
final_animation.save('output.mp4', writer='ffmpeg', fps=10)
final_animation.save('output.gif', writer='imagemagick', fps=10)
```
